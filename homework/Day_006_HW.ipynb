# 檢視與處理 Outliers
### 為何會有 outliers, 常見的 outlier 原因
* 未知值，隨意填補 (約定俗成的代入)，如年齡常見 0,999
* 可能的錯誤紀錄/手誤/系統性錯誤，如某本書在某筆訂單的銷售量 = 1000 本

```python
# Import 需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline

# 設定 data_path
dir_data = './data'
```

```python
# 先篩選數值型的欄位
numeric=app_train.select_dtypes(exclude=object)
# 再把只有 2 值 (通常是 0,1) 的欄位去掉
numeric=numeric.loc[:,numeric.nunique()>2]
print("Numbers of remain columns:%s" % len(list(numeric.columns)))

import seaborn as sns

for col in numeric:
    print(col)
    ax = sns.boxplot(x=numeric[col])    
    ax.set_title(col)
    plt.show()
    print(numeric[col].describe())
    print('--------')
```

### Hints: Emprical Cumulative Density Plot, [ECDF](https://zh.wikipedia.org/wiki/%E7%BB%8F%E9%AA%8C%E5%88%86%E5%B8%83%E5%87%BD%E6%95%B0), [ECDF with Python](https://stackoverflow.com/questions/14006520/ecdf-in-python-without-step-function)

```python
# 最大值離平均與中位數很遠
print(app_train['AMT_INCOME_TOTAL'].describe())

def ecdf(data):
    x=np.sort(data)
    y=np.arange(1,len(x)+1) /len(x)
    return(x,y)

x,y=ecdf(app_train['AMT_INCOME_TOTAL'])

plt.plot(x,y)
plt.xlabel('Value')
plt.ylabel('ECDF')
#plt.xlim([cdf.index.min(), cdf.index.max() * 1.05]) # 限制顯示圖片的範圍
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(x), y)
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()


app_train['AMT_INCOME_TOTAL'].hist()
plt.show()
print(app_train['AMT_INCOME_TOTAL'].value_counts().sort_index(ascending = False))
```

```python
print(app_train['AMT_INCOME_TOTAL'].describe())

# 繪製 Empirical Cumulative Density Plot (ECDF)
"""
YOUR CODE HERE
"""
from statsmodels.distributions.empirical_distribution import ECDF
ecdf = ECDF(app_train['AMT_INCOME_TOTAL'])


plt.plot(ecdf.x, ecdf.y/ecdf.y.max())
plt.xlabel('Value')
plt.ylabel('ECDF')
plt.ylim([-0.05,1.05]) # 限制顯示圖片的範圍

plt.show()

# 改變 y 軸的 Scale, 讓我們可以正常檢視 ECDF
plt.plot(np.log(list(ecdf.x)), ecdf.y/ecdf.y.max())
plt.xlabel('Value (log-scale)')
plt.ylabel('ECDF')

plt.show()


app_train['AMT_INCOME_TOTAL'].hist()
plt.show()
print(app_train['AMT_INCOME_TOTAL'].value_counts().sort_index(ascending = False))
```
